```python
import re
import sys
import pandas as pd
import numpy as np
from Bio import Entrez
from bs4 import BeautifulSoup
from scipy import io, sparse
from sklearn import compute_confusion, plot_confusion
import os
import json

Entrez.email = "u"

def clean_text(text): # text need to be a BeautifulSoup
	try:
		text = text.Abstract.text.replace("\n", "\t")
	except:
		#print(text)
		text = None
	return text
	
def mesh_names(text):
	soup = BeautifulSoup(text, "xml")
	mesh = []
	names = ['DescriptorName', 'QualifierName', 'Keyword', 'descriptorname', 'qualifiername', 'keyword']
	
	for name in names:
		descri = soup.find_all(name)
		for d in descri:
			try:
				if d["MajorTopicYN"] == "Y":
					mesh.append(d.text.lower())
			except:
				pass
			try:
				if d["majortopicyn"] == "Y":
					mesh.append(d.text.lower())
			except:
				pass				
	mesh = list(np.unique(mesh))
	
	return mesh

def make_doc_mesh_matrix(meshs):
	unique = list(np.unique('\n'.join(str(x) for x in meshs).split("\n")))
	
	frames = np.zeros((len(meshs), len(unique)))
	
	for i, mesh in zip(range(len(meshs)), meshs):
		try:
			for m in mesh.split("\n"):
					frames[i][unique.index(m)] = 1
		except:
			pass
		
	df = pd.DataFrame(frames, columns=unique)	
	df.index = np.arange(len(df))
	
	return df

def get_idlist(db, query, retmax=20000):
	handle = Entrez.esearch(db=db,
							sort='relevance',
							retmax=str(retmax),
							retmode='xml',
							term=query) 
							#,usehistory='y')
	res = Entrez.read(handle)
	handle.close()
	
	idlist = res["IdList"]
	return idlist
	
def fetch(db, ids):
	handle = Entrez.efetch(db=db, id=ids, rettype='full', retmode='xml')
	text = handle.read()
	handle.close()
	return text



def db_to_key_word(db):
	if db == "pubmed":
		return "PubmedArticle"
	# TODO for all possible databases
		
	return None

def search(db, query, out_file_name, num_doc):
	idlist = get_idlist(db, query, num_doc)	
	return process_idlist(idlist, db, out_file_name)

def process_idlist(idlist, db, out_file_name):
	meshs = []
	frames = []
	kept = 0
	try:
		df = pd.read_csv(out_file_name, index_col = 0)
		for ids in df["ID"]:
			idlist.remove(str(ids))
		frames.append(df)
		kept = len(df["ID"])
	except:
		pass
	
	start = stop = 0
	
	print(len(idlist))
	
	i = 0	
	while stop != len(idlist):
		print("fetch...")
		stop = min(start+5000, len(idlist))
		texts = fetch(db, idlist[start:stop])
		soup = BeautifulSoup(texts, "xml")
		texts = soup.find_all(db_to_key_word(db))
		for text in texts:
			i += 1

			# Load new ID
			ids = list(filter(lambda x: x['IdType'] == db, text.find_all('ArticleId')))[0].text
			cleaned_text = clean_text(text)

			#print(cleaned_text)
			if i % 1000 == 0:
				print("{} ___ {} ___ {}".format(kept, i, ids))

			if cleaned_text == None or cleaned_text == '':
				continue
			kept += 1

			#mesh = '\n'.join(str(x) for x in mesh_names(text))
			#meshs.append(mesh)
			#print(mesh)

			#frames.append(pd.DataFrame(data={'ID': [ids], 'Text': cleaned_text, 'Mesh': mesh}))
			frames.append(pd.DataFrame(data={'ID': [ids], 'Text': cleaned_text}))

		start = stop

		# Save dataset
		if frames != []:
			df = pd.concat(frames)
			df.index = np.arange(len(df))
			df.to_csv(out_file_name)
		else:
			df = None

	return df



def merge_corpus_noOverlap(type_name_list, pmid_dict, dir_path):

	for iType in range(len(type_name_list)):
		print("\n", type_name_list[iType])
		for jType_str in type_name_list[iType+1:]:
			common_elements = list(set(pmid_dict[type_name_list[iType]]).intersection(set(pmid_dict[jType_str])))
			#print(type_name_list[iType], jType_str, len(common_elements))

			if len(common_elements) > 0:
				pmid_dict[type_name_list[iType]] = [i for i in pmid_dict[type_name_list[iType]] if i not in common_elements]
				pmid_dict[jType_str] = [i for i in pmid_dict[jType_str] if i not in common_elements]

	ids = []; labels = []; iLabel = 0
	for iType in range(len(type_name_list)):
		print("# --> New size of", type_name_list[iType], len(pmid_dict[type_name_list[iType]]))
		ids.extend([str(x) for x in pmid_dict[type_name_list[iType]]])
		labels.extend([iLabel]*len(pmid_dict[type_name_list[iType]]))
		iLabel=1+iLabel
		
	myjson = json.dumps(pmid_dict)
	f = open(os.path.join(dir_path, "merge_confusion_PMID_noOverlap.json"), "w")
	f.write(myjson)
	f.close		

	# Save merged IDs and abstract
	#os.path.join(dir_path, '.'.join(['_'.join(type_name_list), 'csv']))
	process_idlist(ids, "pubmed", os.path.join(dir_path, "{}_raw.csv".format('_'.join(type_name_list))))
	
	# Save corresponding labels
	doc_tuples = list(zip(ids,labels))
	#labels_array = np.array(labels)
	#labels_array = np.reshape(labels_array, (len(labels), 1))
	tmp_file = '_'.join(['_'.join(type_name_list), 'doc'])
	tmp_file = '.'.join([tmp_file, 'csv'])
	#np.savetxt(fname = os.path.join(dir_path, tmp_file), X = labels_array, fmt = '%d')
	#pd.DataFrame(labels_array).to_csv(os.path.join(dir_path, tmp_file))
	pd.DataFrame(doc_tuples, columns=['ID','label']).to_csv(os.path.join(dir_path, tmp_file))
	print("# --> ID lists processed!")

	return



if __name__ == "__main__":
	if sys.argv[1] == "help":
		print("- search <db> <query>|<query1;query2...> <out_csv_file>|<out_dir_path> [<num_doc>|<num_doc1;num_doc2...>] [<merge>]")
		print("\n- mesh_matrix <raw_csv_file> <out_mesh_mat>")



	if sys.argv[1] == "search":
		# 2: db - 3: query|query1;query2... - 4: out_csv_file|out_dir_path
		# 5: num_doc|num_doc1;num_doc2... - 6: merge
		sep = ';'
		num_doc = 20000

		# If single query (ie., only one term without ';')
		# --> single fetch
		if sys.argv[3].find(sep) == -1:
			num_doc = num_doc if len(sys.argv) <= 5 else int(sys.argv[5])
			user_query = ' '.join([sys.argv[3], 'AND (hasabstract[text] AND Humans[Mesh])'])
			df = search(sys.argv[2], user_query, sys.argv[4], num_doc)
		# Else, multiple queries given
		# --> multiple fetch
		else:
			all_queries = sys.argv[3].split(sep)
			all_num_doc = [num_doc]*len(all_queries)

			# Check if multiple doc number
			if len(sys.argv) > 5:
				if not sys.argv[5].find(sep) == -1:
					all_num_doc = [int(x) for x in sys.argv[5].split(sep)]
				else:
					all_num_doc = [int(sys.argv[5])]*len(all_queries)

			# Query pubmed for each query term
			type_name_list = []
			for curr_query, curr_numDoc in list(zip(all_queries, all_num_doc)):
				print(curr_query, curr_numDoc)
				user_query = ' '.join([curr_query, 'AND (hasabstract[text] AND Humans[Mesh])'])
				curr_outFile = curr_query.replace(' ', '-')
				type_name_list.append(curr_outFile)
				curr_outFile = os.path.join(sys.argv[4], '.'.join([curr_outFile, 'csv']))
				df = search(sys.argv[2], user_query, curr_outFile, curr_numDoc)

			# Compute the confusion matrix
			print("# --> Compute the confusion matrix")
			cm, pmid_dict = compute_confusion(type_name_list, sys.argv[4])
			np.save(file = os.path.join(sys.argv[4], "merge_confusion"), arr = cm)
			myjson = json.dumps(pmid_dict)
			f = open(os.path.join(sys.argv[4], "merge_confusion_PMID_overlap.json"), "w")
			f.write(myjson)
			f.close

			# Plot the confusion matrix
			#plot_confusion(cm, type_name_list)

			# Merge the corpus without PMID overlap
			if len(sys.argv) == 7:
				if sys.argv[6] == '1':
					print("# --> Merge the corpus")
					merge_corpus_noOverlap(type_name_list, pmid_dict, sys.argv[4])


	if sys.argv[1] == "retrieve":
		# 2: db - 3: PMID files directory path... - 4: PMID file suffix (*_PMID.txt)

		ids = []; labels = []; iLabel = 0
		for file in os.listdir(sys.argv[3]):
			if file.endswith("_PMID.txt"):
				#print(os.path.join(sys.argv[3], file))
				tmp_pmid_list = np.loadtxt(os.path.join(sys.argv[3], file), dtype=int).tolist()
				tmp_pmid_list = [str(i) for i in tmp_pmid_list]
				ids.extend(tmp_pmid_list)
				labels.extend([iLabel]*len(tmp_pmid_list))
				iLabel = iLabel + 1 

		# Save merged IDs and abstract
		corpus_fileName = '_corpus'
		out_file_name = os.path.join(sys.argv[3], '.'.join([corpus_fileName, 'csv']))
		#print(out_file_name)
		process_idlist(ids, "pubmed", out_file_name)
		
		# Save corresponding labels
		doc_tuples = list(zip(ids,labels))
		tmp_file = '_'.join([corpus_fileName, 'doc'])
		tmp_file = '.'.join([tmp_file, 'csv'])
		pd.DataFrame(doc_tuples, columns=['ID','label']).to_csv(os.path.join(sys.argv[3], tmp_file))
		print("# --> ID lists processed!")
		
		
	if sys.argv[1] == "plotOverlap":
		# 2: merge confusion files directory path...

		# Plot the confusion matrix
		plot_confusion(sys.argv[2])		


	if sys.argv[1] == "mesh_matrix":
		df = pd.read_csv(sys.argv[2], index_col = 0)
		doc_mesh = make_doc_mesh_matrix(df['Mesh'])
		names = doc_mesh.columns
		doc_mesh = sparse.csr_matrix(doc_mesh)
		io.savemat(sys.argv[3], {'X': doc_mesh})
		pd.DataFrame(names).to_csv(sys.argv[3][:-4]+"_vocabulary.csv")
```
